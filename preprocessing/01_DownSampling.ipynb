```python
# Params
sensor = "acc"
f_shape = (64, 1) # @200 Hz
root_dir = "/root/upconversion/data/2018_01_16/"

# name_list = [
#     "01_kawabe", "02_higashinaka", "03_yasuda", "04_kamiya",
#     "05_ishiyama", "06_yoshimura", "07_aiko", "08_higashide",
# ]

# name_list = [
#     #"09_others", "10_others", "11_others", "12_others",
#     #"13_others", "14_others", "15_others",
#     "16_others", "17_others",
# ]
name_list = [
    "30_hada", "31_teramae","32_torisuke", "33_matsukawa","34_sato","35_hamase",
    "36_yamasaki","37_koguchi","38_onuma","39_kashiyama","40_yamaguchi",
    "41_watase","42_oga","43_shigeyoshi","44_hukuda","45_maekawa",
]
```

```python
# write downsampled data
import pandas as pd

cols = ["time", "sub_id", "mod", "sensor", "x", "y", "z"]

for _name in name_list[:]:
    print(">> ", _name)
    df_raw = pd.read_csv(root_dir+"{}/data/Labeled/{}_labeled.csv".format(_name, sensor), index_col=0)
    df_raw["sub_id"] = _name.split("_")[0]
    df_raw["sensor"] = sensor
    df_raw["mod"] = df_raw["time"].str[-1].astype(int)%5
    df_raw["time"] = pd.to_datetime(df_raw["time"], format="%Y%m%d_%H:%M:%S.%f") 
    display(df_raw.head(20))
    for mod in range(5):
        x = df_raw[df_raw["mod"] == mod].loc[:, ["time", "sub_id", "mod", "sensor","x", "y", "z"]]
        #x["mod"] = str(mod)
        
        file_name = "./dataStore_tmp/DownSampled/mg4_200Hz_sub{}_mod{}_{}.csv".format(_name.split("_")[0], mod, sensor)
        x[cols].to_csv(file_name, index=None)
        print(sensor, mod)
        print("File Name: {}, shae={}".format(file_name, x.shape))
    display(x[cols].head())
```
